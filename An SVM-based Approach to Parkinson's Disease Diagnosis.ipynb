```python
from sklearn import svm 
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Parkinsson disease.csv')
```

```python
X = df.drop(columns =['name','status'],axis=1)
Y = df['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
df.shape
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size= 0.1 , random_state=2 )
```

```python
print(X_train.shape ,X.shape)
```

```python
model = StandardScaler()
```

```python
model.fit(X_train)
```

```python
X_train = model.transform(X_train)
X_test = model.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# models training
```

```python
models = svm.SVC(kernel = 'linear')
```

```python
models.fit(X_train,Y_train)
```

```python
# model evaluation and accuracy score 
```

```python
training_data_predicion = models.predict(X_train)
training_data_accuracy = accuracy_score(Y_train , training_data_predicion)
```

```python
print('Accuray_score : ', training_data_accuracy)
```

```python
test_data_predicion = models.predict(X_test)
test_data_accuracy = accuracy_score(Y_test , test_data_predicion)
```

```python
print('Accuracy score of test :  ',test_data_accuracy)
```
